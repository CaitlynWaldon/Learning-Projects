## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
#function to calculate a players power based on their pokemons strength in attack and defense
def power_calc (players,pokedex,player_id):
        p_sum = []
        power = 0
        for p in players[player_id]['player_pokemon'].keys():
            if p in pokedex.keys():
                p_sum.append(pokedex[p]['attack'])
                p_sum.append(pokedex[p]['defense'])
        power = (sum(p_sum))
        print(players[player_id]['player_name']+'s power is '+ str(power))
````

````python
# Check function is working
power_calc(players,pokedex,2)
````

### 9.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?



````python
#create a new list of pokemon totals
totals = []
for p in pd_dict:
    totals.append(float(pd_dict[p]['total']))
    
````

````python
#import numpy and calculate mean and standard deviation of list of pokemon totals
import numpy as np
mean_total = np.mean(totals)
std_total = np.std(totals)
print('mean is ' + str(mean_total))
print('std is ' + str(std_total))
````

### 9.2

The game is no fun if the characters are wildly unbalanced! Are any characters "overpowered", which we'll define as having a "Total" more than three standard deviations from the population mean?

````python
#calculate 3 standard devitations form the mean
power = mean_total + 3*std_total
power2 = mean_total - 3*std_total

#loop through pokedex dictionary and check for any pokemons that are more than 3 standard 
#deviations from the mean
for p in pd_dict:
    if pd_dict[p]['total'] > power:
        print(pd_dict[p]['name'] + ' is overpowered by ' + str((pd_dict[p]['total']-power)))
    elif pd_dict[p]['total'] < power2:
        print(pd_dict[p]['name'] + ' is underpowered by ' + str((pd_dict[p]['total']-power2)))
    
````
