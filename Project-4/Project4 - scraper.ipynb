```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import urllib.request
import time
```

# Web Scraping Indeed

```python
# scrape data science jobs in sydney on Indeed
url_template = 'https://au.indeed.com/jobs?q="data+scientist"&l=Sydney+NSW&start={}'
df = pd.DataFrame(columns=["Title","Location","Company","Salary", "Link", "Description"])
for start in range(0,100,10):
    url = url_template.format(start)
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser', from_encoding="utf-8")
    for each in soup.find_all(class_= "result" ):
        try: 
            title = each.find(class_='jobtitle').text.replace('\n', '')
        except:
            title = 'None'
        try:
            location = each.find('span', {'class':"location" }).text.replace('\n', '')
        except:
            location = each.find('div', {'class':"location" }).text.replace('\n', '')
        
        try: 
            company = each.find(class_='company').text.replace('\n', '')
        except:
            company = 'None'
        try:
             salary = each.find('span', {'class':'no-wrap'}).text.replace('\n', '')
        except:
            salary = 'None'
        try:
            part_link = each.find('div', {'class': 'title'}).a.get('href')
            link = 'https://au.indeed.com' + part_link       
        except:
            link = 'None'
        try:
            html2 = requests.get(link)
            soup2 = BeautifulSoup(html2.content, 'html.parser', from_encoding="utf-8")
            data = soup2.find("div", {"class":"jobsearch-jobDescriptionText"})
            desc = data.text
        except:
            desc = 'None'



        df = df.append({'Title':title, 'Location':location, 'Company':company, 'Salary':salary,'Link':link, 'Description':desc}, ignore_index=True)
        time.sleep(5)
```
